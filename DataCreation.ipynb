```python
# importing libraries
!pip install GoogleNews
!pip install newspaper3k
from GoogleNews import GoogleNews
from newspaper import Article
from newspaper import Config
import pandas as pd
import nltk
```

```python
# configuring browser

user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
config = Config()
config.browser_user_agent = user_agent
```

```python
# searching the autonomous articles

googlenews=GoogleNews(start='01/01/2012',end='04/09/2018')
googlenews.search('Autonomous Cars')
result=googlenews.result()
df=pd.DataFrame(result)
```

```python
# fetching the articles 

for i in range(8,19):
    googlenews.getpage(i)
    result=googlenews.result()
    df=pd.DataFrame(result)
    
```

```python
# removing unavaialble articles
update_df = df.drop([8,47,50,117,200,221])
```

```python
# getting the full article and storing in excel
import nltk
nltk.download('punkt')

listd = []
for i in update_df.index:
    d={}
    article = Article(update_df['link'][i],config=config)
    article.download()
    article.parse()
    article.nlp()
    d['Date']=update_df['date'][i]
    d['Media']=update_df['media'][i]
    d['Title']=article.title
    d['Article']=article.text
    d['Summary']=article.summary
    listd.append(d)
df_news=pd.DataFrame(listd)
df_news.to_excel("articles.xlsx")
```
